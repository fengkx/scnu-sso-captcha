```python
from PIL import Image
from matplotlib.pyplot import imshow
import cv2
imcv = cv2.imread('./test/2bAJ.jpg')
import numpy as np

def cvimshow(cvim):
    img = cv2.cvtColor(cvim, cv2.COLOR_BGR2RGB )
    imshow(img)
```

```python
cvimshow(imcv)
```

```python
gray = cv2.cvtColor(imcv,cv2.COLOR_BGR2GRAY) 
cvimshow(gray)
im2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 15)
cvimshow(im2)
im3 = cv2.medianBlur(im2, 3)
# cvimshow(im3)
im4 = im2.copy()

for i in range(1, im4.shape[0]-1):
    for j in range(1, im4.shape[1]-1):
        nine_arr = np.concatenate((im4[i-1][j-1:j+2],im4[i][j-1:j+2],im4[i+1][j-1:j+2])) / 255
        cnt = np.sum(nine_arr)
        if cnt <= 2.5:
            im4[i][j] = 0
```
